# here we have imported necessory the packages and modules 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
```

# below We are reading from our csv raw data file.

```python
df = pd.read_csv('student_prediction_data.csv')
```

```python
df.info
```

```python
df.head()
```

# Here we decide for supervised learning which label or column will be input for algoritham

```python
X = pd.DataFrame(df.Hours)
```

```python
X.head()
```

# Here we decide for supervised learning which label or column will be output for algoritham

```python
y = pd.DataFrame(df.Scores)
```

```python
y.head()
```

# here we want to devide our data frame or data set into Train data and Test data by using the model_selection module


```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size= .8)
```

```python
len(X_train)
```

# For creating model, we have created object of Linear Regression 

```python
model = LinearRegression()
```

# to train the model

```python
model.fit(X_train, y_train)
```

# checking the accuracy on training data

```python
model.score(X_train, y_train)*100
```

# prediction of test data

```python
predicted_y = model.predict(X_test)
```

```python
predicted_y
```

# here in linear model we actualy calculate the slope and the intercept of the prediction line which is calculated while fitting the model and their values are as follows

```python
coef = model.coef_
```

```python
intercept = model.intercept_
```

# for Visualization of our model and prediction based on our data as follows

```python
plt.scatter(X,y,color = 'green',marker= '*')
x_label = 'Hours'
plt.plot(X_test,predicted_y,color = 'purple')
plt.show()
```
